```python
from detection.detection import Detector
```

```python
detector = Detector(class_name_path='./fish_classes.txt', weights_path='./yolov4_weights_1021_best_3.h5')
```

```python
detector.set_iou_threshold(0.413)
detector.set_score_threshold(0.3)
```

```python
detector.build_model()
detector.load_weights()
```

```python
output_img, detections = detector.detect("./ChaetodonAuriga_003.jpg")
```
